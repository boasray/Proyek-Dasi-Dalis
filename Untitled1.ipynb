```python
from IPython.display import display
import pandas as pd
import re

# Function untuk mendeteksi nama valid
def is_invalid_name(name):
    name = str(name).strip()

    # kosong atau spasi
    if name == "":
        return True

    # hanya angka
    if name.isnumeric():
        return True

    # hanya simbol
    if re.fullmatch(r"[^\w]*", name):
        if not any(c.isalpha() for c in name):
            return True

    return False

# 1. Deteksi nama invalid
invalid_mask = df['Nama'].apply(is_invalid_name)
invalid_names = df[invalid_mask]

print("=== Data yang INVALID pada kolom Nama_Pengguna ===")
display(invalid_names)

# 2. Perbaiki menjadi 'Unknown'
df.loc[invalid_mask, 'Nama'] = "Unknown"

# 3. Tampilkan tabel hasil perbaikan
fixed_rows = df[df['Nama'] == "Unknown"]

print("=== Data setelah diperbaiki (Nama = 'Unknown') ===")
display(fixed_rows)
```

```python
gender = {
    "Laki-laki": 1,
    "Perempuan": 2
}

Usia = {
    "< 18 tahun": 1,
    "18 – 25 tahun": 2,
    "26 – 35 tahun": 3,
    "36 – 45 tahun": 4,
    "> 45 tahun": 5
}

status = {
    "Pelajar/Mahasiswa": 1,
    "Pegawai Negeri Sipil (PNS)": 2,
    "Karyawan Swasta": 3,
    "Wiraswasta": 4,
    "Petani/Nelayan": 5,
    "Lainnya": 6
}

likert = {
    "Sangat Tidak Setuju": 1,
    "Tidak Setuju": 2,
    "Netral": 3,
    "Setuju": 4,
    "Sangat Setuju": 5
}

frekuensi = {
    "Tidak pernah": 1,
    "Jarang": 2,
    "Cukup sering": 3,
    "Sering": 4,
    "Sangat sering": 5
}

durasi_penggunaan = {
    "< 3 jam": 1,
    "3 – 6 jam": 2,           # pastikan pakai tanda –
    "7 – 10 jam": 3,
    "11 – 14 jam": 4,
    "> 14 jam": 5
}

intes_penggunaan = {
    "1 – 3 kali": 1,
    "4 – 6 kali": 2,
    "7 – 9 kali": 3,
    "10 – 12 kali": 4,
    "> 12 kali": 5
}

waktu_intes_penggunaan = {
    "Pagi (06.00 – 11.59)": 1,
    "Siang (12.00 – 17.59)": 2,
    "Malam (18.00 – 23.59)": 3,
    "Larut malam (00.00 – 05.59)": 4,
    "Sepanjang hari (pagi – malam)": 5
}
```

```python
#encode ordinal karena ada urutannya jadi pakai map()
df["rentang usia"] = df["rentang usia"].map(Usia)
```

```python
#encode ordinal untuk colom yang memiliki jawaban pertanyaan tingkat kepuasan
likert_cols = [
    "cek_setelah_bangun",
    "khawatir_baterai_habis",
    "panik_saat_kuota_internet_habis",
    "tidak tenang ketika sinyal hilang",
    "cemas_tidak_bisa_medsos",
    "takut_melewatkan_informasi",
    "pantau_berita_setiap_saat",
    "tidak_nyaman_tanpa_smartphone",
    "bingung_tanpa_smartphone",
    "nyaman_dekat_smartphone",
    "bawa_smartphone_kemana_mana",
    "panik_hp_tertinggal",
    "marah_smartphone_bermasalah",
    "kehilangan_kendali_tanpa_smartphone"
]

for col in likert_cols:
    df[col] = df[col].map(likert)
```

```python
#encode ordinal untuk colom yang memiliki jawaban pertanyaan tingkat kepuasan
frequency_cols = [
    "frekuensi_cek_notifikasi",
    "frekuensi_gunakan_saat_berbicara"
]

for col in frequency_cols :
    df[col] = df[col].map(frekuensi)
```

```python
#encode ordinal karena ada urutannya jadi pakai map()
df["durasi_penggunaan"] = df["durasi_penggunaan"].map(durasi_penggunaan)
```

```python
df = pd.get_dummies(df, columns=["waktu_paling_intens"])

# Ubah hanya kolom hasil one-hot encoding menjadi int
onehot_cols = [col for col in df.columns if col.startswith("waktu_paling_intens_")]
df[onehot_cols] = df[onehot_cols].astype(int)
```

```python
#encode ordinal karena ada urutannya jadi pakai map()
df["frekuensi_buka_smartphone"] = df["frekuensi_buka_smartphone"].map(intes_penggunaan)
```

```python
import pandas as pd
import re

# Hapus isi dalam kurung (...), misalnya "(Facebook, Instagram, dll.)"
df['aktivitas_bersih'] = df['aktivitas_utama'].apply(
    lambda x: re.sub(r"\([^)]*\)", "", x).strip()
)

# Rapikan koma agar hanya 1 spasi setelah koma
df['aktivitas_bersih'] = df['aktivitas_bersih'].str.replace(
    r"\s*,\s*", ", ", regex=True
)

# One-hot encode langsung dimasukkan ke df
df = pd.concat(
    [df, df['aktivitas_bersih'].str.get_dummies(sep=', ')],
    axis=1
)
```

```python
numerical_cols = [
    "durasi_penggunaan",
    "frekuensi_buka_smartphone",
    "frekuensi_cek_notifikasi",
    "frekuensi_gunakan_saat_berbicara"
]

onehot_cols = [
    "Bekerja/Belajar",
    "Belanja online",
    "Bermain Media sosial",
    "Bermain game",
    "Chattingan",
    "Mendengar Hiburan lain",
    "Streaming video"
]

# Ambil kolom waktu paling intens yang namanya diawali dengan prefix ini
waktu_cols = [col for col in df.columns if col.startswith("waktu_paling_intens_")]

# Gabungkan list kolom hasil one-hot encoding aktivitas dengan kolom waktu paling intens
onehot_cols += waktu_cols
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numerical_cols),
        # kolom onehot dilewati (passthrough)
    ],
    remainder='passthrough'
)

X_processed = preprocessor.fit_transform(df[numerical_cols + onehot_cols])
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)  # Membuat 1 komponen utama
X_pca = pca.fit_transform(X_processed)  # Terapkan PCA pada data yang sudah diproses
```

```python
# feature engineering 
df["skor_intensitas_hp"] = (X_pca - X_pca.min()) / (X_pca.max() - X_pca.min()) * 100
```

```python
# feature engineering 
max_score_nomophobia = len(likert_cols) * 5
df["skor_nomophobia"] = (df[likert_cols].sum(axis=1)/max_score_nomophobia)*100
```

```python
corr = df["skor_intensitas_hp"].corr(df["skor_nomophobia"])
print(f"Korelasi Intensitas HP dan Nomophobia: {corr:.3f}")
```

```python
from scipy.stats import spearmanr

corr, p_value = spearmanr(df["skor_intensitas_hp"], df["skor_nomophobia"])
print(f"Korelasi Spearman: {corr:.3f}, p-value: {p_value:.3f}")
```

```python
def kategori_nomophobia(score):
    if score <= 33:
        return "Nomophobia Rendah"
    elif score <= 69:
        return "Nomophobia Sedang"
    else:
        return "Nomophobia Tinggi"

df["kategori_nomophobia"] = df["skor_nomophobia"].apply(kategori_nomophobia)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x="skor_intensitas_hp", y="skor_nomophobia", data=df)
plt.title("Hubungan Skor Intensitas HP dan Nomophobia")
plt.xlabel("Skor Intensitas HP (0-100)")
plt.ylabel("Skor Nomophobia")
plt.show()
```
